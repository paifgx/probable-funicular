```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 6, 9]

plt.plot(x, y, color="red", linestyle="--", marker="o")
plt.plot(x, y2, color="blue", linestyle=":", marker="x")

plt.title("Ein einfaches Liniendiagramm")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")

plt.legend()

plt.show()
```

```python
labels = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(labels, values, color="purple")

plt.title("Ein einfaches Balkendiagramm")
plt.xlabel("Kategorien")
plt.ylabel("Werte")
plt.show()
```

```python
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]

plt.scatter(x, y, color="orange")
plt.title("Ein Streudiagramm")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")
plt.show()
```

```python
data = [1, 2, 2, 2, 3, 4, 4, 5, 5, 6, 7, 8, 9, 9, 9, 10]

plt.hist(data, bins=5, color="teal")
plt.title("Histogramm der Datenverteilung")
plt.xlabel("Werte")
plt.ylabel("Hufigkeit")
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y)
plt.title("Angepasstes Diagramm")
plt.xlabel("X-Achse")
plt.ylabel("Y-Achse")
plt.xlim(0, 5)
plt.ylim(0, 20)
plt.grid(True)

plt.savefig("diagramm.png")
plt.show()
```
